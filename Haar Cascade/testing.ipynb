```python
import cv2
import os
video_src = './videos/MAX_0040_.MP4'
```

```python
import time
import IPython
import csv
```

```python
cap = cv2.VideoCapture(video_src)
```

```python
tree_cascade=cv2.CascadeClassifier("./XML/cascade_final.xml") ##path of cascade file
```

```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
# out = cv2.VideoWriter('output.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))
out = cv2.VideoWriter('output1213.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))
```

```python
# fps = cap.get(cv2.CAP_PROP_FPS)
# delay = int(1000/fps)
```

```python
start = time.time()
print(start)
```

```python
filename = 'timePerSec.csv'
f = open(filename,'w', newline='')
wr = csv.writer(f)
```

```python
i = 0
sum_fps = 0
while True:
    ret, img = cap.read()
#     if (i % 10 != 0) :
#         i+=1
#         continue
    if (type(img) == type(None)):
        break
    # BGR to RGB
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    trees = tree_cascade.detectMultiScale(gray, scaleFactor=1.27, minNeighbors=5, minSize=(200,200))
    
    for (x,y,w,h) in trees:
        # (image, start point, end point, color, thickness)
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),2)
    if cv2.waitKey(1) == 27:
        break

#     if ret == True:
        # it works on local environment
#         cv2.namedWindow('video', cv2.WINDOW_NORMAL)
#         cv2.resizeWindow('video', 1280, 720)
#         cv2.imshow('video', img)
        # Write the frame into the file 'output.avi'
#         out.write(img)
        # delay fps
#         cv2.waitKey(delay)
    
    # calcualte loop time
    end = time.time()
    print(i, ": ", end - start)
    sum_fps += end - start
    wr.writerow([i, float(end - start), i / sum_fps])
    start = end
    i += 1

# When everything done, release the video capture and video write objects
cap.release()
out.release()
print("END")

# cv2.destroyAllWindows()
```
