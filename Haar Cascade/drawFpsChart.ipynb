```python
import matplotlib.pyplot as plt
import pandas as pd
  
  
# Initialize the lists for X and Y
data = pd.read_csv('./timePerSec.csv')
  
df = pd.DataFrame(data)

X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])

# Plot the data using bar() method
plt.bar(X, Y)
plt.title("MAX_0040")
plt.xlabel("frame")
plt.ylabel("sec")
  
# Show the plot
plt.show()
```

```python
A = list(df.iloc[:, 0])
B = list(df.iloc[:, 2])

# Plot the data using bar() method
plt.bar(A, B)
plt.title("MAX_0040")
plt.xlabel("frame")
plt.ylabel("fps")
  
# Show the plot
plt.show()
```
